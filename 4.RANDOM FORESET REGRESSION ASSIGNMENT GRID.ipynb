```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python


dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent=dataset[["charges"]]
```

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
para_grid={'criterion':['mse', 'mae','friedman_mse'],
           'max_features':['sqrt','log2'],
           'n_estimators':[100,1000,2000,3000]}

grid=GridSearchCV(RandomForestRegressor(), para_grid, refit=True, verbose = 3, n_jobs=-1)


grid.fit(independent,dependent)


```

```python
re=grid.cv_results_
```

```python


print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python

table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])# change the paramter,play with it.
print("Future_Prediction={}".format(Future_Prediction))
```
